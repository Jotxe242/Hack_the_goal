```python
s = ("locales", "turistas", "usuarios")
def graficar(gdf, s):
  for i in s:
    ax = gdf.plot(i)
    ax.axis('off')
    #cx.add_basemap(ax, crs=gdf.crs.to_string())
    legendgram(plt.gcf(), # grab the figure, we need it
              ax, # the axis to add the legend
              gdf[i], # the attribute to map
              np.percentile(gdf[i], np.arange(10,110,10)), # the breaks to induce color differences on
              mplpal.Viridis_10 # the palette to use
              )
    plt.show()

graficar(gdf, s)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from legendgram import legendgram
import palettable.matplotlib as mplpal

s = ("locales", "turistas", "usuarios")

fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # Crea una figura con 3 subplots en una fila

# Itera sobre las columnas y crea las gráficas
for i, col in enumerate(s):
    ax = gdf.plot(column=col, ax=axes[i], legend=True)  # Grafica en el subplot correspondiente
    if i != 2:
      axes[i].axis('off')  # Desactiva los ejes
      axes[i].set_title(col)  # Agrega el título al subplot
    else:
      axes[i].axis('off')  # Desactiva los ejes
      axes[i].set_title("usuarios totales")  # Agrega el título al subplot

plt.tight_layout()  # Ajusta el espacio entre los subplots
plt.show()  # Muestra la figura con los subplots
```
